### 데이터 프레임 인덱서 : loc, iloc
- Pandas는 numpy행렬과 같이 쉼표를 사용한 (행 인덱스, 열 인덱스) 형식의 2차원 인덱싱을 지원
    - 특별한 인덱서(indexer) 속성을 제공
    
* loc : 라벨값(문자) 기반의 2차원 인덱싱
* iloc : 순서를 나타내는 정수 기반의 2차원 인덱싱 (0-base 위치 인덱스 사용)


### 인덱서 사용하지 않은 경우

- df[:행] : 행 추출하려면 슬라이싱
- df[열][행]

### iloc 인덱서(위치 인덱스)

- 위치 정수값은 0부터 시작 : 0-base 위치 인덱스 사용
