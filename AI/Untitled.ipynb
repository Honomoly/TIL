## 클래스(class)란?
* 객체를 찍어내는 객체!

어떻게 찍어내지?

```python
a = object() # 호출(call)을 실행
print(a)
```

### 클래스로 찍어낸 객체는 그 클래스의 '인스턴스(instance)'라고 한다

```python
class TestClass:
    def say():
        print('Hello!')
        
    def bye():
        print("See you tomorrow!")
```

```python
a = TestClass()
print(a)
```
