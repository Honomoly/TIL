```python
import pandas as pd
```

```python
df = pd.read_csv('seoul_park.csv')
df.head(2)
```

```python
df.info()
```

```python
df.loc[:, 'X좌표(GRS80TM)':].head()
```

```python
import folium
```

```python
locations = df[['X좌표(WGS84)', 'Y좌표(WGS84)']]
df.iloc[131]
# 131은 결측
```

```python
locations = locations.iloc[:131]
locations.tail()
```

```python
seoul_park = folium.Map(location=[37.5502, 126.982], zoom_start=11)
for idx in locations.index:
    cord = [df.loc[idx, 'Y좌표(WGS84)'], df.loc[idx, 'X좌표(WGS84)']]
    folium.Marker(cord, popup=df.loc[idx, '공원명']).add_to(seoul_park)
seoul_park
```

```python
df_needed = df[['공원명', '지역', '공원주소', 'X좌표(WGS84)', 'Y좌표(WGS84)']]
df_needed.rename({'지역':'자치구', '공원주소':'주소', 'X좌표(WGS84)':'경도', 'Y좌표(WGS84)':'위도'}, axis=1, inplace=True)
df_needed = df_needed.iloc[:131]
df_needed.head()
```

```python
df_needed.to_csv('seoul_park_loc_gu.csv')
```
