## 다중 클래스 분류 연습문제

### 1. 데이터 로드

```python
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
iris_data
```

```python
X_feature = iris_data.data
y_target = iris_data.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_feature, y_target, test_size=0.15, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### 2. 전처리

```python
# 해답 부분만 One-Hot 인코딩
from keras.utils import to_categorical
y_train_pp = to_categorical(y_train)
y_test_pp = to_categorical(y_test)
```

### 3. 모델 생성

```python
from keras.models import Sequential 
from keras.layers import Dense
```

```python
model_iris = Sequential()

model_iris.add(Dense(units=5, input_dim=4, activation='tanh'))
model_iris.add(Dense(units=3, activation='softmax'))

model_iris.compile(loss='categorical_crossentropy', optimizer='adam', metrics='accuracy')
```

```python
history = model_iris.fit(x=X_train, y=y_train_pp,
                         epochs=500, batch_size=4, validation_data=(X_test, y_test_pp))
```

### 4. 모델 저장 / 로드

```python
model_iris.save('../data/model_iris.keras')
```

```python
from keras.saving import load_model
loaded_model = load_model('../data/model_iris.keras')
```

### 5. 시각화

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.title('accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('loss')
plt.legend()

plt.show()
```
