```python
import numpy as np
```

```python
# 준비된 수술 환자 데이터를 불러 오기
Data_set = np.loadtxt("../data/ThoraricSurgery.csv", delimiter=",")
Data_set.shape
Data_set[:3]
```

```python
X_feature = Data_set[:, 1:-1]
y_target = Data_set[:, -1]
X_feature[:3]
y_target[:3]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_feature, y_target, test_size=0.1, random_state=1)
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential # 신경망 모델
from keras.layers import Dense # 신경망 층을 정의
```

```python
# 1
model = Sequential()

# 2
model.add(Dense(units=30, input_dim=16, activation='linear'))
model.add(Dense(units=1, activation='sigmoid'))

# 3
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics='Accuracy')
```

```python
# 4
model.fit(x=X_train, y=y_train, epochs=300, batch_size=5)
```

```python
# 평가
model.evaluate(X_test, y_test)
```
