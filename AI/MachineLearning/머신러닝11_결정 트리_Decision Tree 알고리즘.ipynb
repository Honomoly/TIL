## 결정 트리(Decision Tree)  알고리즘

### 결정 트리 알고리즘
- 학습을 통해 데이터에 있는 규칙을 자동으로 찾아내 트리 기반의 분류 규칙을 만드는 알고리즘
- ML 알고리즘 중 직관적으로 이해하기 쉬운 알고리즘으로 분류와 회귀 문제에서 가장 널리 사용
- 특정 기준(질문)에 따라서 데이터를 구분
- 가장 쉬운 규칙 표현 방법 : if ~ else 기반 

  
### 결정트리 유형
- 분류 트리  
    - 범주형 목표 변수를 기준으로 마디를 나눔
    - 끝마디(말단 노드)에 포함된 자료의 범주가 분류 결과 값이 됨

- 회귀 트리  
    - 연속형 목표 변수를 기준으로 마디를 나눔
    - 끝마디(말단 노드)에 포함된 자료의 평균값이 각 끝마디의 회귀 값이 됨  

### Tree 구조  
- 전체적인 모양이 나무를 뒤집어 놓은 것과 닮았다고 해서 붙여진 이름  
- 결정 트리에서 질문이나 네모 상자를 노드(Node)라고 함  
- 맨 위의 노드(첫 질문)를 Root Node  
- 각 질문에 대한 마지막 노드를 Leaf Node : 결정된 클래스 값 
    - 말단 노드 : 더 이상 분할하지 않는 노드 
        - 규칙 없음 
- 규칙 노드 (Decision Node) : 규칙 조건을 갖고 있는 노드 (이 규칙에 따라 분할)  
- 새로운 규칙 조건만 Sub Tree  생성  

#### 용어
- Root Node : 
    - 모든 데이터를 포함하는 의사결정 트리의 줄발점
    - 데이터를 몇 개의 동일한 그룹으로 분할
- Leaf Node
    - 최종 결과를 나타내는 노드(마디)
    - 말단 노드 (자식 노드 없음)
    - 데이터를 더 이상 분할하지 않음
    - 분할하기 위한 규칙 없음  
- Parent/Child Node
    - 상위 노드 / 하위 노드
- Spliting
    - 주어진 조건에 따라 데이터를 분할하는 작업
- Branch/Sub Tree
    - 트리를 분리하여 생성된 하위 트리
- Depth 
    - Root부터 마지막 노드까지 경로의 길이 
    - 아래 그림에서는 depth=3이 됨

### 결정트리에서 중요한 이슈
- 데이터 세트의 피처가 결합해 규칙 조건을 만들 때마다 규칙노드가 만들어짐  
- 규칙이 많아지면 결정 방식이 복잡해지고 과적합(overfitting) 발생  
    - 즉, depth가 길어질수록 결정트리의 예측 성능이 저하될 가능성이 높음  
- 가능한 적은 결정노드로 높은 예측 정확도를 가지려면  
    - 데이터를 분류할 때 최대한 많은 데이터 세트가 해당 분류에 속할 수 있도록  
    - 결정 노드의 규칙이 정해져야 함  
- 그러기 위해서 어떻게 트리를 분할할 것인가가 중요  
    - 최대한 균일한 데이터 세트를 구성할 수 있도록 분할하는 것이 필요  

### 가지치기 (pruning)  
- 특정 노드 밑의 하부 트리를 제거하여 일반화 성능을 높이는 방법  
- 깊이가 줄어들고 결과의 개수 줄어듦  
- 과적합 발생을 막기 위한 방법  
- (트리에 가지가 너무 많으면 과적합(overfitting) 발생)  

### 결정 트리 장단점  
- 장점
    - 매우 쉽고 유연하게 적용될 수 있는 알고리즘  
    - 정보 균일도 룰을 기반으로 알고리즘이 쉽고 직관적이며 명확함
    - 데이터 스케일링(표준화/정규화) 등의 사전 가공의 영향이 적음  

- 단점   
    - 규칙이 많아지면 결정 방식 복잡해지고 과적합 발생  
    - 즉, depth가 길어질수록 결정트리의 예측 성능이 저하될 가능성이 높음  
    - 전반적으로 결정트리의 성능은 그리 높지 않다

### 균일도 (불순도 : impurity)

#### 균일도  
- 같은 것(값)끼리 섞여 있는 정도  
- 한가지 종류만 있는 경우 균일도 높고  
- 여러 가지가 섞여 있는 경우 균일도 낮고, 혼잡도 높음  
- 규칙 조건 생성에서 중요한 것  
- 정보 균일도가 높은 데이터 세트를 먼저 선택할 수 있도록 규칙 조건을 만든 것이 중요  

#### 균일한 데이터 세트  
- 데이터 세트의 균일도는 데이터를 구분하는 데 필요한 정보의 양에 영향을 미침  

- A : 하얀공, 검은공 비슷하게 섞여 있음. 혼잡도가 높고 균일도가 낮음  
- B : 일부 하얀공, 대부분 검은공  
- C : 모두 검은공. 균일도 가장 높음  
    
C에서 데이터를 추출했을 때 데이터에 대한 별다른 정보 없이도 '검은공'이라고 쉽게 예측 가능  
A의 경우는 상대적으로 혼잡도가 높교 균일도가 낮기 때문에  
같은 조건에서 데이터를 판단하는 데 더 많은 정보 필요  

#### 엔트로피 (Entropy)  
- 데이터 분포의 불순도(impurity)를 수치화한 지표  
- 서로 다른 데이터가 섞여 있으면 엔트로피가 높고  
- 같은 값이 섞여 있으면 엔트로피 낮음  
- 엔트로피를 통해 불순도 판단    
- 수치 0 : 모든 값이 동일 (분류하지 않아도 됨)  
- 수치 1 : 불순도 최대  
    
#### 정보 이득(Information Gain) 지수  
- 분류를 통해 정보에 대해 얻은 지식  
- 1-엔트로피 지수  
- 결정트리는 정보 이득 지수로 분할 기준을 정함  
- 즉, 정보 이득 지수가 높은 속성을 기준으로 분할  

#### 지니 계수  
- 불순도를 수치화한 지표  
    - 경제학에서 불평등 지수를 나타낼 때 사용하는 계수  
    - 0이 가장 평등하고, 1로 갈수록 불평등  
- 머신러닝에 적용될 때는 지니 계수가 낮을수록  
    - 데이터 균일도가 높은 것으로 해석  
    - gini = 0.0 (불순도 0: 한 종류라는 의미)  

#### 결정 트리 알고리즘에서 지니 계수 이용  
- 사이킷런의 DecisionTreeClassifier 클래스는  
- 기본으로 지니 계수를 이용해서 데이터 세트 분할  
- 데이터 세트를 분할하는 데 가장 좋은 조건  
    - 정보 이득이 높거나 지니 계수가 낮은 조건  

#### 결정 트리 알고리즘에서 분류를 결정하는 과정  
- 정보 이득이 높거나 지니 계수가 낮은 조건을 찾아서  
- 자식 트리 노드에 걸쳐 반복적으로 분할한 뒤  
- 데이터가 모두 특정 분류에 속하게 되면  
    - 분할을 멈추고 분류 결정  
    - 이때의 지니 계수는 0

### CART(Classification And Regression Trees) 기반 알고리즘
- 분류뿐만 아니라 회귀도 가능 
- 결정트리, 랜덤 포레트스, GBM, XGBoost, LightGBM 등 트리 기반 알고리즘 
- 트리의 노드마다 이진 분류(Binary classification)하는 특징
- 일반적으로 지니 계수(데이터 군집의 불순도를 측정하는 지표)를 사용
- 지니 계수가 높을수록 군집 내 데이터들의 속성들이 많이 일치  
- 따라서 CART 알고리즘은 지니 계수가 높은 방향으로 Decision Tree의 노드 구성

```python
# 데이터 셋트 로드 및 분리

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
# import warnings
# warnings.filterwarnings('ignore')

# 데이터 세트 로드
iris_data = load_iris()

# 학습/테스트 데이터 세트로 분리
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, 
                                                    iris_data.target, 
                                                    test_size=0.2, 
                                                    random_state=11)
```

### 결정 트리 하이퍼 파라미터 튜닝

#### 결정 트리 하이퍼 파라미터

- 규칙 트리는 규칙 생성 로직을 미리 제어하지 않으면  
- 완벽하게 클래스 값을 구별해 내기 위해  
- 트리 노드를 계속해서 만들어가기 때문에  
- 매우 복잡한 규칙 트리가 만들어져  
- 모델이 쉽게 과적합되는 문제 발생  
- 하이퍼 파라미터를 사용하여 복잡한 트리가 생성되지 않도록 제어 
    - 여러 개의 하이퍼파라미터 값을 조절해서 
    - 트리의 깊이를 조절하여 (더 짦은 트리가 만들어지도록 / 너무 깊은 트리가 생성되지 않도록)  
    - 최적의 성능을 내도록 제어    
    --> 하이퍼 파리머터 튜닝 수행  

#### 결정 트리의 하이퍼 파라미터
- max_depth : 결정 트리의 최대 트리 깊이 제어
- min_samples_split : 자식 규칙 노드를 분할해서 만드는데 필요한 최소 샘플 데이터 개수
    - min_samples_split=4로 설정하는 경우 
    - 최소 샘플 개수가 4개 필요한데
    - 3개만 있는 경우에는 더 이상 자식 규칙 노드를 위한 분할을 하지 않음
    - 트리 깊이도 줄어서 더 간결한 결정 트리 생성
- min_samples_leaf : 리프 노드가 될 수 있는 최소 샘플 개수
    - 리프 노드가 될 수 있는 조건은 디폴트로 1
    - min_samples_leaf=4로 설정하면 
    - 샘플이 최소 4개가 되어야 리프 노드가 된다는 말은
    - 3개, 2개, 1개인 경우에는 리프 노드가 될 수 없다는 것이므로
    - 만약 분할시 샘플 갯수가 4개보다 작아지는 노드가 생겨야할 경우
    - 그 즉시 더 이상의 분할을 진행하지 않음
    - 지니 계수가 크더라도 리프 노드의 샘플 갯수가 4개 이상인 조건으로 규칙 변경되어
    - 브랜치 노드가 줄어들고 결정 트리가 더 간결하게 됨 
- min_samples_split와 min_samples_leaf의 차이
    - 예시) 각각 min_samples_split=4, min_samples_leaf=4인 경우
    - 만약 어느 노드에서 샘플이 5개이고 이 샘플들이 노드의 조건에 의해 2개, 3개로 분류되어야 한다면
    - min_samples_split은 최소 샘플조건(해당 노드에서 4개 이상)을 충족하므로 2개, 3개로 분류가 된다
    - 그러나 min_samples_leaf은 분류시 샘플 갯수가 2개, 3개가 되는 것에서
    - 리프가 될 조건(리프에서 4개 이상)이 충족되지 않으므로 더이상 분류되지 않는다

```python
# max_depth=3
dt_clf = DecisionTreeClassifier(random_state=156, max_depth=3)
dt_clf.fit(X_train, y_train)
plt.figure(figsize=(10, 10))
plot_tree(dt_clf, filled=True, feature_names=iris_feature_names)
```

```python
# min_samples_split=4
dt_clf = DecisionTreeClassifier(random_state=156, min_samples_split=4)
dt_clf.fit(X_train, y_train)
plt.figure(figsize=(10, 10))
plot_tree(dt_clf, filled=True, feature_names=iris_feature_names)
```

```python
# min_samples_leaf=4
dt_clf = DecisionTreeClassifier(random_state=156, min_samples_leaf=4)
dt_clf.fit(X_train, y_train)
plt.figure(figsize=(10, 10))
plot_tree(dt_clf, filled=True, feature_names=iris_feature_names)
```

- 그리드 서치(Grid search)와 유사한 원리이지만

- Grid Search에 비해 불필요한 반복 수행 횟수를 줄일 수 있어 수행 속도 빠름  

### 분류기(dt_clf)의 feature_importance_ 속성

#### feature_importance_ 속성 : 피처 중요도 속성 
- tree를 만드는 결정(분할할 때)에 각 피처가 얼마나 중요한지 평가하는 수치  
- 0~1 사이의 숫자
---
- 결정 트리는 균일도에 기반해서 어떠한 속성(피처)를 규칙 조건으로 선택하느냐가 중요한 요건  
- 중요한 몇 개의 피처가 명확한 규칙 트리를 만드는데 크게 기여하며
- 모델을 좀 더 간결하고 이상치에 강한 모델을 만들 수 있음
---
feature_importances_
- ndarray 형태로 값을 반환
- 피처 순서대로 값이 할당됨
- [첫 번째 피처의 중요도, 두 번째 피처의 중요도, ....]
- 값이 높을 수록 해당 피처의 중요도가 높다는 의미
- 특정 노드의 중요도 값이 클수록, 그 노드에서 불순도가 크게 감소 의미

```python
# 피처별 중요도
pd.Series(dt_clf.feature_importances_, index=iris_feature_names)
```

### 결정 트리 과적합 (Overfitting)

#### 결정 트리 과적합 (Overfitting)
- 과적합 : ML에서 학습 데이터를 과하게 학습하는 것
- 과적합 문제 : 학습 데이터에서는 정확하지만 테스트 데이터에서는 성과가 나쁜 현상
- 의사결정 트리 과적합 : 나무의 크기가 크거나 가지의 수가 많을 때 과적합 문제 발생
---
- 과소적합 (Underfitting)
    - 학습이 제대로 이루어지지 않은 상태
    - 데이터가 불충분하거나 모델이 지나치게 단순하여 학습량이 충분치 않을 경우  
---

- 다양한 데이터에 대해 적용할 수 있는 일반적인 모델을 만드는 것
