```python
import pymysql
```

```python
# 1. DB 연결
# 각 파라미터들에 정확한 정보 입력
conn = pymysql.connect(host='localhost',
                       port=3306,
                       user='root',
                       passwd='Wk0qbfmt!',
                       charset='utf8')
```

```python
# 2. cursor 객체 생성
cursor = conn.cursor()
```

```python
# 3. sql문 실행
sql = 'SHOW DATABASES'
cursor.execute(sql)
```

```python
# 4. sql문 실행결과 추출 / fetchall() 메소드
result = cursor.fetchall()
result
```

```python
# 새로운 데이터베이스 만들기
sql = 'CREATE DATABASE beauty_shop'
cursor.execute(sql)
cursor.fetchall()
```

```python
sql = 'DROP DATABASE beauty_shop'
cursor.execute(sql)
cursor.fetchall()
```

```python
# 데이터베이스 사용
sql = 'USE sqldb3'
cursor.execute(sql)
```

```python
sql = 'SHOW TABLES'
cursor.execute(sql)
```

```python
result = cursor.fetchall()
result
```

```python
# 테이블 구조 확인
sql = 'DESC book'
cursor.execute(sql)
cursor.fetchall()
```
