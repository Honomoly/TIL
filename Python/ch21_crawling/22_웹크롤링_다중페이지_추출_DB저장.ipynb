```python
def get_request_product(url):
    html = urlopen(url)
    bs_obj = bs4.BeautifulSoup(html, 'html.parser')
    return bs_obj
```

```python
def get_product_info(box):
    strong_tag = box.select_one('strong.name')
    li_tag = box.findAll('li')
    
    name = strong_tag.text.split(':')[1]
    price = li_tag[0].text.replace('\nPrice', '').strip()
    sale_price = li_tag[-1].text.strip()
    return {'í’ˆëª©':name, 'ì •ìƒê°€':price, 'í• ì¸ê°€':sale_price}
```

```python
def get_page_product(url):
    global all_product_df
    bs_obj = get_request_product(url)
    boxes = bs_obj.select('div.description')
    
    # ì²« í˜ì´ì§€ë§Œ 22ê°œ, ë‚˜ë¨¸ì§€ëŠ” 20ê°œ
    if url.split('=')[1] != '1':
        boxes = boxes[2:]
        
    for box in boxes:
        df = pd.DataFrame(get_product_info(box), index=range(1,2))
        all_product_df = pd.concat([all_product_df, df], axis=0, ignore_index=True)
    return all_product_df
```

```python
# í…Œì´ë¸” ìƒì„±
sql = """
    CREATE TABLE product (
        prdNo INT AUTO_INCREMENT PRIMARY KEY,
        prdName VARCHAR(2000),
        prdPrice FLOAT,
        prdDisPrice FLOAT
    )
"""

cursor.execute(sql)
```

```python
# save_data() ì‘ì„±
def save_data(prd_info):
    sql = "INSERT INTO product (prdName, prdPrice, prdDisPrice) VALUES (%s, %s, %s)"
    values = (prd_info['prdName'], prd_info['prdPrice'], prd_info['prdDisPrice'])
    cursor.execute(sql, values)
```

```python
# ê° ë°•ìŠ¤ì—ì„œ ì •ë³´ë§Œ ìˆ˜ì •í•´ì„œ ì¶”ì¶œ
def get_product_info(box):
    try:
        name = box.find('strong', {'class':'name'}).text.split(':')[1]
        # ì˜¤ë¥˜ê°€ ë‚˜ëŠ” ë¬¸ì ì œê±°
        name = name.replace("'", "")
        name = name.replace("ğŸ”¸", "")
        
        price = box.find('ul').findAll('span')[1].text
        sale_price = box.find('ul').findAll('span')[-1].text
        
        # ì„¸ì¼ê°€ê²©ì´ ì—†ìœ¼ë©´ ì—†ìŒ ì²˜ë¦¬
        if sale_price == "":
            sale_price = "USD 0.0"
            
    except:
        print("ì¶”ì¶œì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤.")
    
    return {'prdName' : name, 'prdPrice' : price[4:], 'prdDisPrice' : sale_price[4:]}
```

```python
def save_page_product(url):
    html = urlopen(url)
    bs_obj = bs4.BeautifulSoup(html, 'html.parser')
    boxes = bs_obj.select('div.description')
    
    # ì²« í˜ì´ì§€ë§Œ 22ê°œ, ë‚˜ë¨¸ì§€ëŠ” 20ê°œ
    if url.split('=')[1] != '1':
        boxes = boxes[2:]
        
    for box in boxes:
        save_data(get_product_info(box))
```

```python
base_url = 'https://jolse.com/category/toners-mists/1019/?page='
for i in range(1, 23):
    print('%2dë²ˆì§¸ í˜ì´ì§€ ì¶”ì¶œì¤‘'%i)
    url = base_url + str(i)
    save_page_product(url)
conn.commit()
```
