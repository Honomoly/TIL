### 서울의 모든 약국

```python
endpoint = 'http://apis.data.go.kr/B552657/ErmctInsttInfoInqireService/'
oper = 'getParmacyListInfoInqire?'
QT = '1'
Q0 = quote('서울특별시')
ORD = 'NAME'
pageNo = '1'
numOfRows = '50'

param = oper + \
        'serviceKey=' + serviceKey + \
        '&QT=' + QT + \
        '&Q0=' + Q0 + \
        '&ORD=' + ORD + \
        '&pageNo=' + pageNo + \
        '&numOfRows=' + numOfRows
url = endpoint + param
```

```python
result = requests.get(url)
bs_obj = bs4.BeautifulSoup(result.content, 'html.parser')
bs_obj
```

```python
names = bs_obj.findAll('dutyname')
tels = bs_obj.findAll('dutytel1')
addrs = bs_obj.findAll('dutyaddr')
namesl, telsl, addrsl = [[] for _ in range(3)]
for n, t, a in zip(names, tels, addrs):
    namesl.append(n.text)
    telsl.append(t.text)
    addrsl.append(a.text)

df = pd.DataFrame({'약국명':namesl, '전화번호':telsl, '주소':addrsl})
df
```

```python
# 월요일 밤 9시 이후 영업하는 약국
names, tels, addrs = [[] for _ in range(3)]
all_items = bs_obj.findAll('item')
for i, item in enumerate(all_items):
    time = int(item.find('dutytime1c').text)
    if time>=2100:
        names.append(item.find('dutyname').text)
        tels.append(item.find('dutytel1').text)
        addrs.append(item.find('dutyaddr').text)
        
df2 = pd.DataFrame({'약국명':names, '전화번호':tels, '주소':addrs})
df2
```
